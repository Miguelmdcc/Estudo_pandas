```python
product = 'Soja'
milho_mean_price_date = mercado_dados.query('@product not in Product')
print(milho_mean_price_date.Product.unique())
milho_mean_price_date = milho_mean_price_date.groupby('Date')[['Price']].mean().sort_values('Price')
milho_mean_price_date.plot(kind='barh',figsize=(14,10),color='blue',title='Preço médio por data do milho')
```

```python
product_dados = mercado_dados.Product.value_counts(normalize=True).to_frame().sort_values('Product',ascending=True)
product_dados
```

```python
product_dados.plot(kind='bar',figsize=(14,10),color='blue',title='Percentual das quantidade de Produtos',xlabel='Products',ylabel='Percentual')
```
